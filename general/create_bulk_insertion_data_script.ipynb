```python
# importing libraries
import pandas as pd
import json
```

```python
# reading reviews file
df = pd.read_csv('D:\\ADB\\sf_csv\\reviews_1.csv', delimiter = '~')
df.head()
```

```python
# creating copy of original dataframe
newdf = df.copy(deep = True)
newdf.head()
```

```python
# dropping current review ids
newdf = newdf.drop('review_id', axis=1)
newdf
```

Generating review_ids for:

Insertion Set 1: contains reviews with 1 million reviews

Insertion Set 2: contains reviews with 5 million reviews

Insertion Set 3: contains reviews with 10 million reviews

Insertion Set 4: contains reviews with 20 million reviews

```python
# 1 million bulk insertion
reviewdf_1Mil = newdf.loc[newdf.index.repeat(1000)].reset_index(drop=True)

reviewIDData = []
for i in range(1000000):
    reviewIDData.append('sample_review_id'+str(i+1))
print(len(reviewIDData))

assert(len(reviewIDData) == 1000000)

reviewdf_1Mil["review_id"] = reviewIDData

reviewdf_1Mil.to_json('insertBulk_1Mil.json', orient = 'records', lines = True)

print(reviewdf_1Mil.head())
```
