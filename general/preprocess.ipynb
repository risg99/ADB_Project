```python
import json
import numpy as np
import pandas as pd
```

```python
def load_json_df(file):
    data_file = open(f"data/{file}.json", encoding= 'utf8')
    data = []
    for line in data_file:
        data.append(json.loads(line))
    data_file.close()
    df = pd.DataFrame(data)
    return df
```

```python
business = load_json_df('business')
```

```python
user = load_json_df('user')
```

```python
review = load_json_df('review')
```

```python
print('\nBusinesses\n')
display(business.head(3))
print('\nUsers\n')
display(user.head(3))
print('\nReviews\n')
display(review.head(3))
```

```python
def generate_subsets():
    n_subset = [1000, 10000, 100000, 1000000, 2000000]
    for i in range(len(n_subset)):
        reviews = review.iloc[:n_subset[i]]
        users = user.loc[user.user_id.isin(reviews.user_id)].reset_index(drop = True)
        businesses = business.loc[business.business_id.isin(reviews.business_id)].reset_index(drop = True)
        
        reviews.to_json(f'sf_data/reviews_{i+1}.json', orient = 'records', lines = True)
        print(f'Reviews with SF {i+1} has been loaded')
        users.to_json(f'sf_data/users_{i+1}.json', orient = 'records', lines = True)
        print(f'Users with SF {i+1} has been loaded')
        businesses.to_json(f'sf_data/businesses_{i+1}.json', orient = 'records', lines = True)
        print(f'Businesses with SF {i+1} has been loaded\n')
```

```python
generate_subsets()
```
